## 03.용어

**말뭉치 Corpus (collection of texts)**
- 자연어 처리에서 분석하거나 학습에 사용하는 텍스트 데이터의 모음이다.
- 특정 주제의 뉴스 기사 모음, 소셜 미디어 게시글 모음, 법률 문서 등.
- 일반적으로 특정 도메인에 따라 구성되며, 크고 다양한 데이터를 포함하는 것이 이상적이다.
- 예전에는 품사, 형태소 등의 보조적 의미를 추가하여 구조적인 형태로 정리해 놓은 것 포함했었다.

**토큰 Token**
- 자연어 문서를 분석하기 위해 긴 문자열을 작은 단위로 나눈 것.
- 텍스트를 분석하기 위해 나누는 기본 단위로, 일반적으로 단어, 문자, 구두점 등이 포함될 수 있다.
- 문장 "I love NLP!"의 토큰은 ["I", "love", "NLP", "!"]로 나눌 수 있다.
- Tokenize: 문자열을 여러 개의 조각, 즉 여러 개의 Token(토큰, 단어)들로 쪼개는 것을 말한다.
- 특수 token: `<START>`, `<EOS>`, `<UNK>`, `<PAD>`, etc

**문장 Sentence**
- 자연어 텍스트에서 문장부호나 문법적 단위에 따라 나눠지는 문맥의 단위이다. (Sequence of words)
- "NLP is fascinating. I enjoy learning it."는 두 개의 문장으로 나뉜다.
- 문장 분리(Sentence Segmentation)는 일반적으로 구두점을 기준으로 하지만, 특정 언어에서 더 복잡할 수 있다.

**단어 Words**
- 문장에서 의미를 가지는 최소 단위이다.(Sequence of meaningful characters)
- "I love NLP."의 단어는 ["I", "love", "NLP"]이다.
- 띄어쓰기나 언어 규칙에 따라 나뉜다.
    - 영어의 경우 공백나 구두점(,;) 으로 구분할 수 있다.   
    - 일부 언어(중국어, 일본어 등)는 띄어쓰기가 없어 추가 처리가 필요하다
    - 한국어, 일본어의 경우는 보통 공백 없이 문장을 작성해도 단어단위로 읽을 수 있다. (한국어에 띄어쓰기가 적용되기 시작한것은 1900년대 이후이다.)

**형태소 Morpheme**
- 의미를 가지는 가장 작은 언어 단위로, 단어보다 더 세부적으로 나뉠 수 있다.
- 영어의 "unbreakable"은 [un-, break, -able]로 나뉜다.
    - un-: 접두사(부정 의미)
    - break: 어근
    - -able: 접미사(가능 의미)
- 형태소 분석(Morphological Analysis)을 통해 어근, 접사 등을 파악할 수 있다. 한국어와 같이 형태소가 중요한 언어에서 많이 사용된다. 한글 형태소 분석기나 Sub-word 기법이 사용된다.

**불용어 Stop-words**
- 불용어. 자주 등장하지만 분석을 하는 것에 있어서는 큰 도움이 되지 않는 단어들
- 예를 들면, I, my, me, over, 조사, 접미사 같은 단어들은 문장에서는 자주 등장하지만 실제 의미 분석을 하는데는 거의 기여하는 바가 없다.
- NLTK에서는 위와 같은 100여개 이상의 영어 단어들을 불용어로 패키지 내에서 미리 정의하고 있다.



## 04. Text Vectorization

텍스트 벡터화(Text Vectorization)는 복잡한 언어 데이터를 컴퓨터가 이해할 수 있는 **숫자의 집합(벡터)** 로 바꾸는 과정이다.

### 1. 텍스트 벡터화란

텍스트 벡터화는 크게
"무엇을 벡터로 만드느냐"와
"어떻게 의미를 담느냐"에 따라 다음과 같이 구분한다.

**A. 표현 단위에 따른 분류**

1. 문서 표현 (Document Representation)
   문장이나 문서 전체를 하나의 벡터로 표현한다.
   (예: 이 메일은 스팸인가?)

2. 단어 표현 (Word Representation)
   개별 단어를 각각의 벡터로 표현한다.
   (예: '사과'와 '배'의 유사도 측정)

**B. 표현 방식에 따른 분류**

1. 국소 표현 (Local Representation)
   단어 자체만 보고 숫자를 부여한다.
   문맥이나 의미 차이를 무시하는 경향이 있다.
   (Sparse Vector / 희소 표현)

2. 연속 표현 (Continuous Representation)
   주변 단어와의 관계를 통해 의미를 학습한다.
   숫자가 꽉 찬 형태를 띤다.
   (Dense Vector / 밀집 표현)


### 2. 핵심 기법 비교: 희소 표현 vs 밀집 표현

| 구분    | 희소 표현 (Sparse)                | 밀집 표현 (Dense)                   |
| ----- | ----------------------------- | ------------------------------- |
| 주요 기법 | One-hot encoding, BoW, TF-IDF | Word2Vec, GloVe, FastText, BERT |
| 특징    | 대부분의 값이 0인 고차원 벡터             | 실수로 가득 찬 저차원 벡터                 |
| 의미 반영 | 단순 빈도나 존재 여부만 기록              | 단어 간의 유사도와 의미를 반영               |
| 단점    | 단어가 많아질수록 차원이 폭발              | 학습에 상대적으로 많은 자원 필요              |


### 3. 기술의 진화 단계 (Roadmap)

**1단계: 단순 카운트 (One-hot, BoW)**

* 단어가 "있다/없다" 혹은 "몇 번 나왔다"만 체크한다.
* '사과'와 '포도'가 과일이라는 연관성을 전혀 알지 못한다.


**2단계: 통계적 중요도 (TF-IDF)**

* 단순 빈도가 아니라, 전체 문서에서 얼마나 희귀하고 중요한지를 계산한다.
* 단어의 특징을 잡아내기 시작하지만, 여전히 의미(Context)는 모른다.


**3단계: 의미 학습 (Word2Vec, FastText)**

* "비슷한 위치에 등장하는 단어는 의미도 비슷하다"는 가정하에 학습한다.
* `왕 - 남자 + 여자 = 여왕`과 같은 벡터 연산이 가능해진다.


**4단계: 문맥의 완성 (BERT, Transformer)**

* 같은 단어라도 문맥에 따라 뜻이 달라짐을 이해한다.
* 예: "금융 은행"과 "강가의 은행나무"의 '은행'을 서로 다른 벡터로 처리한다.
