# n-gram Language Model

가장 단순한 언어 모델 중 하나인 n-gram 언어 모델을 학습한다.

n-gram 언어 모델은 신경망을 사용하지 않는다. 대신 학습 데이터에서 단어가 얼마나 자주 함께 등장했는지 세고, 그 빈도를 바탕으로 다음 단어의 확률을 계산한다.

예를 들어 학습 데이터에서 `오늘은` 다음에 `날씨가`가 자주 나왔다면, 모델은 `오늘은` 다음 단어로 `날씨가`가 나올 확률을 높게 본다.

## 1. n-gram이란?

n-gram은 연속된 n개의 토큰 묶음이다.

- unigram: 토큰 1개 단위
- bigram: 연속된 토큰 2개 단위
- trigram: 연속된 토큰 3개 단위

예를 들어 문장이 다음과 같다고 하자.

```text
나는 학교에 간다
```

bigram은 다음과 같다.

```text
(나는, 학교에), (학교에, 간다)
```

bigram 언어 모델은 바로 이전 단어 1개만 보고 다음 단어의 확률을 계산한다.

$$
P(w_i \mid w_{i-1})
$$

즉, “이전 단어가 주어졌을 때 다음 단어가 무엇일 가능성이 높은가?”를 빈도 기반으로 계산한다.


````python
import nltk
from nltk.util import ngrams
from collections import Counter

# 예제 말뭉치
text = "오늘은 날씨가 좋다. 오늘은 기분이 좋다. 오늘은 일이 많다. 오늘은 사람이 많다. 오늘은 날씨가 맑다. 어제는 기분이 좋았다."

tokens = nltk.word_tokenize(text)

print(tokens)
````

````python
# unigram : 토큰 1개 단위
unigram = tokens

# bigram : 연속 된 토큰 2개 단위
bigram = list(ngrams(tokens, 2))

print('unigram 예시 : ', unigram)
print('bigram 예시 : ', bigram)
````

````python
unigram_counts = Counter(unigram)
bigram_counts = Counter(bigram)

print('unigram 빈도')
print(unigram_counts)
print()

print('bigram 빈도')
print(bigram_counts)
````

## 2. bigram 확률 계산

bigram 확률은 다음과 같이 계산한다.

$$
P(w_2 \mid w_1) = 
\frac{count(w_1, w_2)}{count(w_1)}
$$

예를 들어 `오늘은`이 5번 등장했고, 그중 `오늘은 날씨가`가 2번 등장했다면 다음과 같다.

$$
P(날씨가 \mid 오늘은) = 
\frac{2}{5}
$$

즉, `오늘은` 다음에 `날씨가`가 나올 확률은 0.4이다.

````python
# bigram 확률 계산
for (w1, w2), bigram_count in bigram_counts.items():
    prob = bigram_count / unigram_counts[w1]
    print(f'P({w2}|{w1}) = {prob:.4f}')
````

````python
# 특정 단어 다음에 올 수 있는 후보만 따로 확인
current_word = '오늘은'

candidates = []
for (w1, w2), count in bigram_counts.items():
    if w1 == current_word:
        prob = count / unigram_counts[w1]
        candidates.append((w2, count, prob))

for word, count, prob in candidates:
    print(f"{current_word} -> {word} : count={count}, prob={prob}")
````

## 3. Perplexity

Perplexity는 언어 모델이 문장을 얼마나 잘 예측하는지 평가하는 지표이다.

쉽게 말하면, Perplexity는 모델이 문장을 보고 얼마나 “당황하는지”를 나타내는 값으로 볼 수 있다.

- 낮은 Perplexity: 모델이 문장을 자연스럽게 예측함
- 높은 Perplexity: 모델이 문장을 낯설게 느낌

문장 $W = w_1, w_2, \ldots, w_N$에 대해 Perplexity는 다음과 같이 표현할 수 있다.

$$
\text{Perplexity}(W) = 2^{-\frac{1}{N}\sum_{i=1}^{N}\log_2 P(w_i \mid w_{i-1})}
$$


1. $P(w_i \mid w_{i-1})$

   현재 단어 $w_i$가 이전 단어 $w_{i-1}$ 다음에 나올 확률이다.

   예를 들어 `오늘은` 다음에 `날씨가`가 나올 확률이 0.4라면 다음과 같다.

   $$
   P(날씨가 \mid 오늘은) = 0.4
   $$

2. $\log_2 P(w_i \mid w_{i-1})$

   확률값에 로그를 취한 것이다.

   문장은 여러 단어의 확률을 곱해서 계산해야 하는데, 확률을 계속 곱하면 값이 너무 작아진다.  
   그래서 실제 계산에서는 확률을 곱하는 대신 로그를 더하는 방식을 사용한다.

3. $\sum_{i=1}^{N}$

   문장 안에 있는 모든 단어에 대해 로그 확률을 더한다는 의미이다.

   즉, 문장 전체가 얼마나 자연스럽게 예측되었는지를 계산한다.

4. $\frac{1}{N}$

   문장 길이가 길수록 로그 확률의 합이 커지므로, 단어 개수 $N$으로 나누어 평균을 구한다.

5. $-$

   로그 확률은 보통 음수로 나오기 때문에, 앞에 마이너스를 붙여 양수 값으로 바꾼다.

6. $2^{(\cdot)}$

   로그로 계산한 값을 다시 원래 확률 척도에 가까운 형태로 되돌린다.

결국 Perplexity는 다음과 같이 이해하면 된다.

- 모델이 각 단어를 높은 확률로 잘 예측하면 Perplexity가 낮아진다.
- 모델이 각 단어를 낮은 확률로 어렵게 예측하면 Perplexity가 높아진다.

예를 들어 어떤 문장에서 다음 단어를 예측할 때마다 모델이 평균적으로 2개 정도의 후보 중에서 헷갈린다면 Perplexity는 2에 가깝다.  
반대로 평균적으로 10개 정도의 후보 중에서 헷갈린다면 Perplexity는 10에 가깝다.

따라서 Perplexity는 작을수록 좋은 값이다.

여기서는 bigram 모델을 사용하므로 이전 단어 1개만 보고 다음 단어의 확률을 계산한다.

````python
import math

def compute_perplexity(sentence, unigram_counts, bigram_counts):
    tokens = nltk.word_tokenize(sentence)
    bigrams = list(ngrams(tokens, 2))
    if len(bigrams) == 0:
        return None
    
    log_prob_sum = 0
    # bigram -> bigrams 로 수정
    N = len(bigrams)

    for w1, w2 in bigrams:
        # 학습 데이터에서 (w1, w2)가 몇 번 등장했는지 확인
        bigram_count = bigram_counts.get((w1, w2), 0)
        # 학습 데이터에서 w1가 몇번 등장했는지 확인
        unigram_count = unigram_counts.get(w1, 0)
        # 학습 데이터에 없는 조합이면 확률이 0이 되고 log(0) 은 연산 불가하므로 작은 값으로 대체
        if bigram_count == 0 or unigram_count == 0:
            prob = 1e-10
        else:
            prob = bigram_count / unigram_count
        
        log_prob_sum += math.log2(prob)

    cross_entropy = -log_prob_sum / N
    perplexity = 2 ** cross_entropy

    return perplexity
````

````python
# 학습 데이터와 비슷한 문장, 낯선 문장을 함께 평가해본다.
test_text = "오늘은 날씨가 좋다. 오늘은 짬뽕 좋더라. 내일보자. 오늘은 기분이 좋다."

for sent in nltk.sent_tokenize(test_text):
    pp = compute_perplexity(sent, unigram_counts, bigram_counts)
    print(f'{sent} : {pp}')
````

## 4. 라플라스 스무딩

n-gram 모델의 가장 큰 문제 중 하나는 학습 데이터에 없는 단어 조합의 확률이 0이 된다는 점이다.

예를 들어 학습 데이터에 `오늘은 짬뽕`이라는 bigram이 한 번도 등장하지 않았다면 다음 확률은 0이 된다.

$$
P(짬뽕 \mid 오늘은) = 0
$$

확률이 0이면 문장 전체 확률도 0이 되고, Perplexity 계산도 불안정해진다.

라플라스 스무딩은 모든 bigram 빈도에 1을 더해서, 처음 보는 조합에도 아주 작은 확률을 부여하는 방법이다.

$$
P(w_2 \mid w_1) =
\frac{count(w_1, w_2) + 1}{count(w_1) + V}
$$

여기서 $V$는 vocabulary size, 즉 전체 단어 종류 수이다.

````python
def compute_perplexity_with_laplace_smoothing(sentence, unigram_counts, bigram_counts):
    tokens = nltk.word_tokenize(sentence)
    bigrams = list(ngrams(tokens, 2))
    if len(bigrams) == 0:
        return None
    
    log_prob_sum = 0
    N = len(bigrams)
    vocab_size = len(unigram_counts)

    for w1, w2 in bigrams:
        bigram_count = bigram_counts.get((w1, w2), 0)
        unigram_count = unigram_counts.get(w1, 0)
        # 라플라스 스무딩 적용
        prob = (bigram_count + 1)/(unigram_count + vocab_size)

        log_prob_sum += math.log2(prob)

    cross_entropy = -log_prob_sum / N
    perplexity = 2 ** cross_entropy

    return perplexity
````

````python
# 스무딩 전/후 비교
for sent in nltk.sent_tokenize(test_text):
    pp = compute_perplexity(sent, unigram_counts,bigram_counts)
    pp_smoothing = compute_perplexity_with_laplace_smoothing(sent,unigram_counts,bigram_counts)

    print('문장 : ',sent)
    print(f'스무딩 전 perplexity: {pp}')
    print(f'스무딩 후 perplexity : {pp_smoothing}')
    print()
````

## 5. n-gram 기반 다음 단어 예측

bigram 빈도를 이용하면 특정 단어 다음에 올 단어를 샘플링할 수 있다.

예를 들어 `오늘은` 다음에 나온 단어들의 빈도를 확인한 뒤, 빈도가 높은 단어가 더 잘 뽑히도록 선택한다.


````python
import random

def sample_next_word(current_word):
    """
    현재 단어 다음에 올 단어를 bigram 빈도 기반으로 샘플링하는 함수
    """

    candidates = []
    for (w1,w2), count in bigram_counts.items():
        if w1 == current_word:
            candidates.append((w2,count))
    
    # 후보가 없으면 문자을 더 이어갈 수 없다
    if len(candidates) == 0:
        return None
    
    words, counts = zip(*candidates)
    totals = sum(counts)

    # 빈도가 높을수록 선택할 확률이 높아진다.
    probs = [count/totals for count in counts]

    return random.choices(words, weights=probs, k=1)[0]

print(sample_next_word('오늘은'))
````

````python
def generate_text_bigram(seed_word, max_len=0):
    """
    bigram 모델을 사용해 간단한 텍스트를 생성하는 함수
    """

    current_word = seed_word
    generate_text = [current_word]

    for _ in range(max_len - 1):
        next_word = sample_next_word(current_word)

        # 더 이상 이어갈 후보가 없으면 생성 중단
        if next_word is None:
            break

        generate_text.append(next_word)
        current_word = next_word

    return ' '.join(generate_text)

print(generate_text_bigram('오늘은', max_len=10))
print(generate_text_bigram('어제는', max_len=10))
````
