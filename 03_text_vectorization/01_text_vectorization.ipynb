# Text Vectorization

텍스트 데이터를 머신러닝이나 딥러닝 모델이 처리할 수 있도록 수치 벡터 형태로 변환하는 기법이다.

텍스트는 그대로는 모델이 이해할 수 없기 때문에, 결국 숫자로 바꾸는 과정이 필요하다.  
이 과정을 텍스트 벡터화라고 한다.

## 텍스트 벡터화의 큰 분류

### 1. 문서 표현 (Document Representation)
텍스트를 문서 단위로 하나의 벡터로 표현하는 방식이다.

- CountVectorizer: 단어 빈도 기반
- TfidfVectorizer: 단어 중요도 기반
- Doc2Vec: 문서의 의미를 학습

### 2. 단어 표현 (Word Representation)
텍스트를 단어 단위로 벡터화하는 방식이다.

- One-Hot Encoding: 단어를 희소 벡터로 표현
- Word Embedding: 단어를 밀집 벡터로 표현
  - Word2Vec
  - GloVe
  - FastText
  - BERT Embedding

---

## 1. Local Representation (국소 표현)

단어를 각각 독립적으로 표현하는 방식이다.  
기본적으로 단어의 의미 유사성은 잘 반영하지 못하며, 문맥 정보도 제한적이다.

### 1-1. One-hot Vector

- 각 단어를 고유한 벡터로 표현
- 예: `[0, 0, 1, 0, 0]`
- 단어마다 하나의 위치만 1이고, 나머지는 모두 0
- 단어 간 유사도를 반영하지 못함

### 1-2. N-gram

- 연속된 토큰 묶음을 표현하는 방식
- 예: 2-gram → "나는", "는 밥", 3-gram → "나는 밥"
- 단어 순서나 일부 국소 문맥을 반영할 수 있음
- 하지만 차원이 쉽게 커지고 희소성이 심해질 수 있음

### 1-3. Count Based (BoW, DTM)

- 특정 단어가 문서 안에서 몇 번 등장했는지 세는 방식
- 대표 예:
  - Bag of Words (BoW)
  - Document-Term Matrix (DTM)
- 단어 빈도 정보는 반영하지만, 일반적으로 단어 의미나 문맥은 잘 반영하지 못함

---

## 2. Continuous Representation (연속적 표현)

단어를 연속된 실수 벡터 형태로 표현하는 방식이다.  
단순히 단어를 구분하는 것을 넘어서, 단어 간 의미적 유사성이나 문맥 정보를 더 잘 반영하려는 목적을 가진다.

### 2-1. Prediction Based (예측 기반)

주변 단어를 예측하거나, 중심 단어를 예측하는 학습 방식으로 벡터를 학습한다.

- 예:
  - Word2Vec
  - FastText

#### Word2Vec
- 대표적인 단어 임베딩 기법
- CBOW / Skip-gram 방식 사용
- 비슷한 문맥에 자주 등장하는 단어가 비슷한 벡터를 갖도록 학습

#### FastText
- Word2Vec의 확장 방식
- 단어를 내부 subword 단위까지 나누어 학습
- OOV(Out Of Vocabulary) 단어 처리에 강함
- 신조어, 오탈자, 형태 변화가 있는 단어에 비교적 유리

### 2-2. Count Based (통계 기반 의미 표현)

단순 빈도 세기에 그치지 않고, 공기(co-occurrence) 정보나 행렬 분해 같은 통계 기법을 활용해 의미를 반영하는 방식이다.

- 예:
  - LSA
  - GloVe

#### LSA (Latent Semantic Analysis)
- 전체 문서를 기준으로 DTM을 만든 뒤 SVD로 차원 축소
- 단어와 문서의 잠재 의미 구조를 추출
- 문서 전체 수준의 통계적 의미를 반영

#### GloVe (Global Vectors for Word Representation)
- 단어쌍 공기 행렬을 기반으로 학습
- Word2Vec보다 전역적인 통계 정보를 더 잘 반영
- count 기반과 embedding 방식의 성격을 함께 가지는 대표 기법

### 2-3. Contextual Embedding (문맥 기반 임베딩)

같은 단어라도 문장 안에서 어떤 문맥에 놓였는지에 따라 다른 벡터로 표현하는 방식이다.

- 예:
  - BERT Embedding

#### BERT Embedding
- Transformer 기반 사전학습 모델
- 문장의 양방향 문맥을 반영
- 같은 단어라도 문맥에 따라 다른 벡터를 생성할 수 있음

예:
- "나는 은행에 갔다"의 "은행"
- "은행에서 일한다"의 "은행"

위 두 문장에서 "은행"은 같은 단어이지만, BERT는 서로 다른 의미로 표현할 수 있다.

특징:
- 문맥과 의미를 가장 잘 반영하는 편
- 단어 벡터뿐 아니라 문장 벡터 수준으로도 활용 가능
- 연산 비용이 큼

## 다양한 벡터화 기법

### 1. BOW (Bag of Words)

문서에 등장한 단어를 문맥이나 순서를 무시하고, 각 단어의 빈도만으로 표현하는 방식이다.

<img src="https://miro.medium.com/v2/resize:fit:1400/1*S8uW62e3AbYw3gnylm4eDg.png" width="500px">

### 2. One-Hot Encoding

단어를 고유 인덱스로 변환한 뒤, 해당 위치만 1이고 나머지는 0인 벡터로 표현하는 방식이다.

<img src="https://d.pr/i/5nWhHo+"/>

### 3. Word Embedding

단어를 밀집 벡터(dense vector)로 표현하는 방법이다.  
단어의 의미와 관계를 보존하면서 벡터 공간에 배치하려는 방식이다.

<img src="https://miro.medium.com/v2/resize:fit:1400/1*jpnKO5X0Ii8PVdQYFO2z1Q.png" width="500px">

| Word          | 성별   | 귀족   | 나이   | 음식   |
|---------------|--------|--------|--------|--------|
| Man (5391)    | -1.07  | 0.01   | 0.03   | 0.04   |
| Woman (9853)  | 1.34   | 0.02   | 0.02   | 0.01   |
| King (4914)   | -0.95  | 0.93   | 0.70   | 0.02   |
| Queen (7157)  | 0.97   | 0.95   | 0.69   | 0.01   |
| Apple (456)   | 0.00   | -0.01  | 0.03   | 0.95   |
| Orange (6257) | 0.01   | 0.00   | -0.02  | 0.97   |

### 4. FastText

Word2Vec의 확장 방식으로, 단어 내부의 subword 정보까지 고려한다.

- 예: 단어를 더 작은 단위로 나누어 학습
- OOV 단어 처리에 강함
- 신조어나 오탈자, 형태 변화가 있는 단어에도 비교적 유리함

### 5. BERT Embedding

문장의 문맥을 양방향으로 고려하는 사전학습 기반 벡터화 방식이다.  
연산 비용은 크지만, 문맥과 의미를 매우 잘 반영한다.

- "나는 은행에 갔다" vs "은행에서 일한다"의 "은행"을 다르게 표현 가능
- 문장 전체 의미를 반영한 단어 벡터, 문장 벡터 생성 가능
