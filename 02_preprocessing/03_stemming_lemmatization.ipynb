# 어간 추출(Stemming) 및 표제어 추출(Lemmatization)
텍스트 데이터에서 서로 다른 형태의 단어들이 같은 의미를 가진다면, 이를 하나의 단어로 일반화하여 분석할 수 있다.

특히 BoW(Bag of Words)와 같이 단어의 빈도를 기반으로 하는 분석에서 매우 유용하게 사용된다.

**단어의 빈도를 이용한 분석에서 어간/표제어 추출**

1. 의미 일관성 : 동의어 처리로 같은 단어로 정규화
2. 데이터 차원 축소 : 차원의 수가 감소하여 계산 비용 감소
3. 노이즈 감소 : 다양한 형태의 단어를 모델이 잘못된 패턴으로 학습할 수 있는 위험 제거
4. 정확성 향상 : 정확한 빈도 계산 가능.


**어간 (Stem)** : 단어의 의미를 담고 있는 단어의 핵심부분

**표제어(Lemma)** : 표제어는 단어의 사전적 형태로, 문법적으로 올바른 형태를 가진 단어이다. 즉 단어의 기본형으로, 언어의 문법적 규칙에 따라 변형된 단어를 원래의 형태로 돌려놓는 과정.

```python
import nltk
# WordNetLemmatizer
nltk.download('wordnet')
```

## 어간추출
단어의 핵심의미를 가진 어간(stem)을 남기는 방법. 단순 규칙에 따라 어미를 잘라내므로, 실제 존재하지 않는 어간이 나올수 있음

- running, runs, ran -> run
- fairness, fair -> fair
- swiftly -> swiftli (이런 단어 없음)

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# 어간 추출기
stemmar = PorterStemmer()
sentence = 'The runners were running swiftly and easily. They ran past the finish line.'

tokens = word_tokenize(sentence)
print(tokens)

stemmed_tokens = [stemmar.stem(token) for token in tokens]
print(stemmed_tokens)
```

## 표제어추출
단어의 원형을 찾아내고, 실제 사전상 원형으로 복구한다. 이때 품사 지정이 필요하다.

**WordNet품사표**

| 품사 태그 | 의미       | 예시                      |
| ----- | -------- | ----------------------- |
| `'n'` | 명사(Noun) | apple, dog, house       |
| `'v'` | 동사(Verb) | run, eat, play          |
| `'a'` | 형용사(Adj) | big, smart, blue        |
| `'r'` | 부사(Adv)  | quickly, silently, well |

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# 동사 표제어 추출 
vbs = ['running', 'ran', 'runs', 'have', 'has', 'am', 'is', 'are', 'were']
vbs = [lemmatizer.lemmatize(vb,pos='v') for vb in vbs]
print(vbs)

# 형용사 표제어 추출
adjs = ['better', 'worse', 'worst', 'faster', 'slower']
adjs = [lemmatizer.lemmatize(adj,pos='a') for adj in adjs]
print(adjs)
```

```python
# nltk 품사를 wordnet 품사로 변호ㅘㄴ
from nltk.corpus import wordnet
from nltk import pos_tag

def pos_nltk_to_wordnet(pos_tag):
    """nltk pos_tag를 매개인자로 받는 함수"""
    if pos_tag.startswith('N'):
        return wordnet.NOUN     # n
    if pos_tag.startswith('V'):
        return wordnet.VERB     # v
    if pos_tag.startswith('J'):
        return wordnet.ADJ      # a
    if pos_tag.startswith('R'):
        return wordnet.ADV      # r
    else:
        return None

def lemmatize(text):
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    lemmatized_tokens = []
    for token, pos in pos_tags:
        wordnet_pos = pos_nltk_to_wordnet(pos) or wordnet.NOUN
        lemma = lemmatizer.lemmatize(token,pos=wordnet_pos)
        lemmatized_tokens.append(lemma)
    return lemmatized_tokens

text = "The runners were running happily and better than others."
print(text)

# 표제어 추출
lemmatized_tokens = lemmatize(text)
lemmatized_tokens
```
