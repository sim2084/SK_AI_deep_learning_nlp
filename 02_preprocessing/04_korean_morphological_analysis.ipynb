# 한국어 형태소분석 및 어간 추출
한국어의 경우, 동사와 형용사는 어간과 어미로 구성된다. 어간 추출을 위해서 어미를 규칙적으로 제거하는 과정이 필요하다.

- 예 : "달리다" -> "달리-:+ "-다"

한국어 어간 추출 예제 : 한국어 어간 추출은 형태소 분석기를 사용하여 더 정교하게 수행이 가능하다.

- 한국어 품사 구조
  1. 체언 : 명사, 대명사, 수사
  2. 수식언 : 관형사, 부사
  3. 관계언 : 조사
  4. 독립언 : 감탄사
  5. 용언 : 동사, 형용사

위와 같이 한국어는 5언 9품사의 구조를 갖는다.

> 어간(stem)

용언(동사, 형용사)을 활용할 때, 원칙적으로 모양이 변하지 않는 부분을 활용해서 어미에 선행하는 부분, 때론 어간의 모양도 변경할 수 있다. (긋다. 긋고, 그어서, 그어라)

> 어미(ending)

용언의 어간 뒤에 붙여서 활용하면서 변하는 부분이며, 여러 문법적 기능을 수행

## KoNLPY의 형태소 분석기
한국어 자연어 처리를 위한 파이썬 라이브러리로 이 라이브러리는 한국어 텍스트 분석에 필요한 다양한 기능을 제공하며, 형태소 분석, 품사 태깅, 텍스트 전처리 등을 손쉽게 수행할 수 있도록한다.

KoNLPy는 여러 형태소 분석기를 통합하여 사용자가 적합한 분석기를 선택할 수 있게 해준다.

**한국어 형태소 분석기 비교**

| 분석기      | 특징 요약                                                                 | 장점                          | 단점                     | 주요 활용/특이점                       |
|-------------|--------------------------------------------------------------------------|-------------------------------|--------------------------|----------------------------------------|
| Hannanum    | 다양한 분석 후보 반환, 품사 태그 세분화(9/22단계 선택)                   | 후보 다양성, 태그 상세 선택   | 속도 느림                | ntags로 품사 세분화 조절 가능           |

| Okt(Twitter)| 트위터 기반, 정규화/어간 추출 지원, 속도 빠름, 명사 추출에 강점           | 정규화, 명사/어구 추출 강점   | 품사 태그 단순, 덜 정교함| 소셜미디어, 빠른 분석, 전처리에 적합     |

### Hannanum

* 다양한 분석 후보 제공, 품사 태그 세분화(9/22단계 선택) 가능.
* 분석 결과의 세밀한 조정이 필요할 때 유리하다.

### Kkma

* 품사 태그 종류가 매우 다양, 문장 분리 기능 지원.
* 정확도 높으나 속도가 느려 대용량 처리에는 부적합.
* 세밀한 품사 정보가 필요할 때 적합하다.

### Mecab

* 일본어 기반이지만 한국어 사전 적용 가능.
* 분석 속도가 매우 빠르고, 대용량 데이터에 적합.
* 사전 커스터마이즈가 필요할 수 있다.

### Komoran

* 자소 단위, 확률 기반 분석, 속도와 성능의 균형.
* 불규칙 어절 처리에 강점, 사전 관리가 용이하다.

### Okt (Open Korean Text)

* 트위터에서 개발, 정규화(norm), 어간 추출(stem) 기능 제공.
* 속도 빠르고 명사/어구 추출에 강점.
* 품사 태그가 단순해 정교한 분석에는 한계가 있다.

**공통 기능**

* 모두 명사 추출(nouns), 형태소 추출(morphs), 품사 태깅(pos) 함수 제공.

**속도 비교**

* 가장 빠름: Mecab > Komoran > Okt > Kkma
* 대용량 데이터, 실시간 처리에는 Mecab, Komoran 추천.

**정리**

* 정확도/품사 다양성: Kkma, Hannanum
* 속도/경량화: Mecab, Komoran
* SNS/비정형 데이터, 명사 추출: Okt

## Okt
Open korean text

```python
from konlpy.tag import Okt

okt = Okt()
text = "아버지가 방에 들어가신다."

# 형태소 분석
tokens = okt.morphs(text)
print(tokens)

# 품사 태깅
pos_tags = okt.pos(text)
print(pos_tags)

# 명사 추출(키워드 추출)
nouns = okt.nouns(text)
print(nouns)
```

```python
# 형태소 분석 - 어간 추출(동사/형용사 등의 사전적 원형을 반환)
text = '나는 버스를 기다리고 있었다.'

tokens = okt.morphs(text,stem=False)
print(tokens)

tokens = okt.morphs(text,stem=True)
print(tokens)
```

## komoran

```python
from konlpy.tag import Komoran

komoran = Komoran()
text = "아버지가 방에 들어가신다."

# 형태소 분석
tokens = komoran.morphs(text)
print(tokens)

# 품사 태깅
pos_tags = komoran.pos(text)
print(pos_tags)

# 명사 추출
nouns = komoran.nouns(text)
print(nouns)
```
