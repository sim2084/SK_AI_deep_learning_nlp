# One-hot Encoding

### 1. 원-핫 인코딩이란?

단어를 벡터로 표현하는 기본 방식이다. 단어 집합 크기만큼 길이를 가진 벡터에서 해당 단어 위치만 1이고 나머지는 0으로 표현한다.

예시: `{'apple': 0, 'banana': 1, 'cherry': 2}` → `cherry` → `[0, 0, 1]`

### 2. 장점

* 직관적이고 구현이 간단하다 - 이해하기 쉽고 빠르게 적용 가능
* 수치 안정성이 뛰어나다 - 0과 1만 사용하므로 계산 오류가 적다
* 범주형 데이터에 최적화 - 순서가 없는 데이터(성별, 색상 등)를 올바르게 표현
* 실무에서 광범위하게 사용 - 표 형식 데이터 분석의 표준 방식

### 3. 한계점과 해결 방안

| 한계점             | 해결 방안                             |
| --------------- | --------------------------------- |
| 카테고리가 많으면 차원 증가 | 상위 카테고리만 선택하거나 희귀 범주 통합           |
| 의미 유사성을 반영하지 못함 | NLP는 워드 임베딩 활용, 범주형 데이터는 원-핫으로 충분 |

### 4. 언제 사용할까?

* 범주형 변수 (성별, 국가, 제품 타입)
* 카테고리가 많지 않을 때 (<50개)
* 신경망 모델 - 범주형 입력을 수치 벡터로 변환하기 위해 사용
* 선형 모델 (로지스틱 회귀, 선형 회귀) - 범주 간 순서 관계를 제거하기 위해 필요. 라벨 인코딩을 사용하면 모델이 범주를 순서가 있는 연속형 변수로 잘못 해석할 수 있다.

## one-hot encoding 예제

```python
# 토큰화
from nltk.tokenize import word_tokenize

setences = "I love NLP so much and I love you"

tokens = word_tokenize(setences)
print(tokens)
```

```python
# 정수 인코딩
from collections import Counter

# 단어 빈도 사전
word_counts = Counter(tokens)
print(word_counts)

# 단어 사전
word_to_index = {word: i for i, word in enumerate(word_counts)}
print(word_to_index)

# 정수 인덱스 변환
sequence = [word_to_index[word] for word in tokens]
print(sequence)
```

```python
# 원 핫 인코딩
import numpy as np

# 단어 사전 수, 총 클래스 수 확인
vocab_size = len(word_to_index)
print(vocab_size)

def one_hot_encoding(sequence, vocab_size):
    one_hot_vecs = []

    for seq in sequence:
        one_hot_vec = [0] * vocab_size
        one_hot_vec[seq] = 1
        one_hot_vecs.append(one_hot_vec)

    return np.array(one_hot_vecs)

one_hot_vectors = one_hot_encoding(sequence,vocab_size)
print(one_hot_vectors.shape)        # (문장의 길이,단어수)
one_hot_vectors
```

## 어린왕자 예제

```python
# 어린왕자 샘플 텍스트
raw_text = """The Little Prince, written by Antoine de Saint-Exupéry, is a poetic tale about a young prince who travels from his home planet to Earth. The story begins with a pilot stranded in the Sahara Desert after his plane crashes. While trying to fix his plane, he meets a mysterious young boy, the Little Prince.

The Little Prince comes from a small asteroid called B-612, where he lives alone with a rose that he loves deeply. He recounts his journey to the pilot, describing his visits to several other planets. Each planet is inhabited by a different character, such as a king, a vain man, a drunkard, a businessman, a geographer, and a fox. Through these encounters, the Prince learns valuable lessons about love, responsibility, and the nature of adult behavior.

On Earth, the Little Prince meets various creatures, including a fox, who teaches him about relationships and the importance of taming, which means building ties with others. The fox's famous line, "You become responsible, forever, for what you have tamed," resonates with the Prince's feelings for his rose.

Ultimately, the Little Prince realizes that the essence of life is often invisible and can only be seen with the heart. After sharing his wisdom with the pilot, he prepares to return to his asteroid and his beloved rose. The story concludes with the pilot reflecting on the lessons learned from the Little Prince and the enduring impact of their friendship.

The narrative is a beautifully simple yet profound exploration of love, loss, and the importance of seeing beyond the surface of things."""
```

```python
# 토큰화 + 전처리
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

sentences = sent_tokenize(raw_text)
en_stopwords = set(stopwords.words('english'))
preprocessed_sentences = []

for sent in sentences:
    tokens = word_tokenize(sent)

    tokens = [
        token.lower()
        for token in tokens
        if token.lower() not in en_stopwords
        and token.isalpha()
        and len(token) > 2
    ]

    preprocessed_sentences.append(tokens)

print(preprocessed_sentences)
```

```python
# 정수 인코딩 + 패딩
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

vocab_size = 30
tokenizer = Tokenizer(num_words=vocab_size + 1, oov_token='OOV')
tokenizer.fit_on_texts(preprocessed_sentences)
sequences = tokenizer.texts_to_sequences(preprocessed_sentences)
print(sequences)

sequences = pad_sequences(
    sequences,
    padding='pre',
    truncating='pre',
    maxlen=16
)
print(sequences)
```

```python
# %pip install torch
```

```python
# 원핫 인코딩 변환
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# 현재 squences의 shape : (문장 수, 문장 길이)(13,16)
X = torch.tensor(sequences, dtype=torch.long)

# 각 정수 토큰을 one-hot 벡터로 변환
# 0은 padding에 사용 되고 실제 단어 인덱스들이 그 뒤를 사용하므로 +1
input_dim = vocab_size+1
X = F.one_hot(X,num_classes=input_dim).float()

# 결과 shape : (배치 크기, 문장 길이, 원-핫 차원 수)
X.shape
```

```python
# X : 원-핫 인코딩 된 문장 데이터
# y : 각 문장의 정답 라벨
y = torch.tensor([0,0,0,1,0,1,2,2,2,2,0,2,2])
```

```python
# 모델 생성
# RNN 모델은 순서가 있는 데이터를 처리할 때 자주 사용한다.
class RNNTextClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes):
        super().__init__()

        self.rnn = nn.RNN(
            input_size=input_dim,           # 한 시점에 들어오는 단어 벡터 길이 (one-hot 차원 수)
            hidden_size=hidden_dim,         # RNN이 내부적으로 유지할 은닉 상태 차원 수
            batch_first=True                # 입력 shape로 (batch_size, seq_len, input_dim) 형태를 사용하도록 설정
        )

        # 마지막 은닉 상태 -> 최종 클래스 점수(logit) 변환
        self.fc = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        # output : 모든 시점의 은닉 상태
        # h_n : 마지막 시점의 은닉상태
        output, h_n = self.rnn(x)

        # h_n shape는 (num_layer,batch_size,hidden_dim)
        # 현재는 1층이 RNN이므로 squeeze 후 (batch_size,hidden_dim) 형태로 변경하여 처리
        logits = self.fc(h_n.squeeze(0))
        return logits
    
# 입력 데이터
model = RNNTextClassifier(input_dim=input_dim,hidden_dim=16,num_classes=3)

# 손실 함수 (다중 분류용)
criterion = nn.CrossEntropyLoss()

# 옵티마이저
optimizer = optim.Adam(model.parameters(),lr=0.005)
```

```python
# 모델 학습
for epoch in range(50):
    optimizer.zero_grad()

    output = model(X)
    loss = criterion(output,y)
    loss.backward()
    optimizer.step()

    if (epoch+1)%10==0:
        print(f'epoch {epoch+1} : Loss{loss.item()}')
```

```python
# 모델 예측
model.eval()

with torch.no_grad():
    output = model(X)
    pred = output.argmax(dim=1)
    print('실제 라벨 : ',y.tolist())
    print('예측 라벨 : ',pred.tolist())
    
```
