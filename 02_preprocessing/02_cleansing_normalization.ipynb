# 정제 Cleansing | 정규화 Normalization

정규화 작업은 토큰화 작업에 방해가 되는 부분을 배제하고 토큰화 작업을 수행하기 위해서 토큰화 작업보다 먼저 진행되기도 하지만 토큰화 이후에 남아있는 노이즈들을 제거하기위해 진행하기도한다.

완벽한 정규화는 사실상 어려운 편이기 때문에 목적에 따라 어느정도 타협을 해야하는 경우가 많다.

**주요 목적**

- 정제(Cleaning) : 불필요한 특수 문자, 이모지, 구두점 등 텍스트 데이터에 포함된 노이즈(Noise) 요소를 제거하여 순수한 정보로 만든다.
- 정규화(Normalization) : 같은 의미지만 표기가 다른 단어들을 일관된 표현으로 통합한다. 이를 통해 중복 표현을 줄이고, 검색 및 분석의 정확성을 높인다.

## 정규화

```python
# United Kingdom, UK -> UK로 통합
text = "The United Kingdom and UK Have a long history together."
normalized_text = text.replace('United Kingdom','UK')
normalized_text
```

```python
# 대소문자 통합
text = 'Theater attendance has decreased. The THEATER industry is adapting!'
normalized_text = text.lower()
normalized_text
```

## 정제
의미가 다소 적고, 불필요한 단어 등을 제거해서 텍스트(corpus)의 품질을 높일 수 있다.

### 등장빈도수가 낮은 단어 제거

```python
# nltk 데이터 셋
import nltk

# 텍스트 토큰화
nltk.download('punkt')
nltk.download('punkt_tab')

# 감성 분석 도구 사전 데이터
nltk.download('vader_lexicon')

# 품사 태깅
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from collections import Counter
from nltk.tokenize import word_tokenize

text = "The quick brown fox jumps over the lazy dog. The fox is quick and agile."
tokens = word_tokenize(text)
word_counter = Counter(tokens)

print(word_counter)

# 등장 빈도가 2회 이상인 단어만 필터링
filtered_tokens = [token for token in word_counter if word_counter[token] >= 2]
filtered_tokens
```

### 짧은 단어 제거 (영어)
통상 영어단어는 6\~7개, 한국단어는 2\~3개정도로 영어에서는 짧은 단어를 제거하는 것만으로 불필요한 토큰수를 줄이고, 텍스트처리의 효율을 높일수 있다.

```python
text = 'A fox is on the run in the forest.'

tokens = word_tokenize(text)
filtered_tokens = [token for token in tokens if len(token) > 2]
filtered_tokens
```

### 노이즈 제거
실제 활용한 내용이 아닌 모든 텍스트를 가리킴.
- 개인정보(이메일/url/전화번호...)
- 웹크롤링된 텍스트에서 태그정보 제거

```python
# 정규 표현식을 이용한 노이즈 제거
import re
text = 'Contact us at master@lifeisgood.com for more details. Visit our site at https://lifeisgood.com'

email_pattern = r'\S+@\S+'          # r : raw 이스케이핑이 아니라 문자로 인식
url_pattern = r'https\S+'

text = re.sub(email_pattern,'',text)
text = re.sub(url_pattern,'',text)
text
```

### 불용어 stopwords 제거
텍스트 데이터에 자주 등장하지만 분석에 있어 큰 의미가 없는 단어.
the, a, an, is, am, 은, 는, 이, 가 ...

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

en_stopwords = stopwords.words('english')
print(en_stopwords)

text = 'She sells seashells by the seashore, and it is a beautiful day.'

tokens = word_tokenize(text)
filtered_tokens = [token for token in tokens if token.lower() not in en_stopwords]
filtered_tokens
```

#### 한국어 불용어제거

```python
from konlpy.tag import Okt

text = "이 방법은 특히 빅데이터에서 중요한 역할을 합니다. 이를 통해 더 많은 정보를 얻을 수 있습니다."

ko_stopwords = set('은 는 이 가 을 를 에서 도 수 이를 통해 더 합니다 입니다 있습니다 없습니다'.split())

# 한국어 형태소 분석기
okt = Okt()

tokens = okt.morphs(text)
print(tokens)

filtered_tokens = [token for  token in tokens if token not in ko_stopwords]
filtered_tokens
```

#### 사용자 정의 불용어 사전 활용

```python
def load_stopwords(filepath):
    with open(filepath,'r',encoding='utf-8') as f:
        stopwords = [line.strip() for line in f]
        return stopwords
```

```python
# https://github.com/stopwords-iso/stopwords-ko
ko_stopwords = load_stopwords('stopwords-ko.txt')

tokens = okt.morphs(text)
filtered_tokens = [token for token in tokens if token not in ko_stopwords]
filtered_tokens
```
