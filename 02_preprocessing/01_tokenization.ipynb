# Tokenization

문장이나 단어를 더 작은 단위로 나누어 분석 가능한 단위(토큰, Token)로 변환하는 과정이다.

토큰의 단위가 상황에 따라 다르지만, 보통 의미있는 혹은 처리하는 단위로서 토큰을 정의한다.

## nltk

NLTK(자연어 처리 툴킷, Natural Language Toolkit)는 파이썬 프로그래밍 언어로 작성된 자연어 처리(NLP) 라이브러리이다. NLTK는 텍스트 처리, 문법 분석, 언어 모델링, 텍스트 분류, 구문 분석 등을 위한 다양한 도구와 데이터셋을 제공한다.

### 주요 기능

1. 토큰화(Tokenization): 문장을 단어 또는 문장 단위로 나누는 작업이다. 예를 들어, "I love NLP."를 ['I', 'love', 'NLP', '.']와 같이 나누는 기능을 제공한다.
2. 품사 태깅(Part-of-Speech Tagging): 각 단어에 대해 해당 품사를 태깅하는 작업이다. 예를 들어, "I love NLP."에 대해 [('I', 'PRP'), ('love', 'VBP'), ('NLP', 'NNP'), ('.', '.')]와 같이 태깅한다.
3. 명사구 추출(Chunking): 문장에서 명사구와 같은 특정 구문을 추출하는 작업이다.
4. 어근 추출(Lemmatization) 및 어간 추출(Stemming): 단어의 기본 형태를 찾는 작업으로, 동사의 기본형을 찾거나 복수형을 단수형으로 변환하는 등의 작업을 한다.
5. 텍스트 분류(Classification): 분류 모델을 사용해 텍스트를 분류할 수 있다.
6. 코퍼스(corpus) 제공: 영화 리뷰, 뉴스 기사 등 여러 텍스트 데이터셋을 포함하고 있어 학습과 실습에 유용하다.

```python
# nltk 데이터 셋
import nltk

# 텍스트 토큰화
nltk.download('punkt')
nltk.download('punkt_tab')

# 감성 분석 도구 사전 데이터
nltk.download('vader_lexicon')

# 품사 태깅
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

## 토큰화

**토큰화 주의사항**

토큰화 작업을 단순하게 코퍼스에서 구두점을 제외하고 공백 기준으로 잘라내는 작업이라고 간주할 수는 없는데 이러한 일은 보다 섬세한 알고리즘이 필요하다.

1. 구두점이나 특수 문자를 단순 제외해서는 안 된다.

   * 코퍼스에서 단어들을 걸러낼 때, 구두점이나 특수 문자를 단순히 제외하는 것은 옳지 않다. 예를 들어, 마침표(.)는 문장의 경계를 알 수 있게 해주므로 단어를 뽑아낼 때 마침표를 제외하지 않을 수 있다.
   * 또 다른 예로, 단어 자체에 구두점을 포함하고 있는 경우가 있다. 예를 들어, m.p.h(시속 마일), Ph.D(박사 학위), AT&T(통신사)와 같이 특정 의미를 갖는 경우가 있다. 특수 문자 또한 $45.55와 같은 가격을 나타내거나, 01/02/06와 같은 날짜를 표현하는 데 사용될 수 있다. 이런 경우, 45.55를 하나의 토큰으로 취급하고 45와 55로 분리하고 싶지 않을 수 있다.

2. 줄임말과 단어 내에 띄어쓰기가 있는 경우.

   * 토큰화 작업에서 영어권 언어의 아포스트로피(')는 축약된 단어를 다시 펼치는 역할을 하기도 한다. 예를 들어, "what're"는 "what are"의 줄임말이며, "we're"는 "we are"의 줄임말이다. 이러한 형태는 접어(clitic)라고 하며, 단어가 줄임말로 쓰일 때 생기는 형태이다.
   * "New York"이나 "rock 'n' roll"과 같은 단어는 중간에 띄어쓰기가 있지만 하나의 단어로 인식해야 할 때가 있다. 따라서, 토큰화 작업은 이러한 단어를 하나의 토큰으로 인식할 수 있는 능력이 필요하다.

3. 표준 토큰화

* Penn Treebank Tokenization의 규칙이 적용된 토큰화: `word_tokenize`, `TreebankWordTokenizer`
* 규칙 1: 하이픈으로 구성된 단어는 하나로 유지한다.
* 규칙 2: "doesn't"와 같이 아포스트로피가 있는 단어는 분리한다.

```python
text = "NLP is fascinating. It has many applications in real-world scenarios."
```

### subword 토큰화

**하위단어 수준 및 글자 수준 토큰화(Subword & Character-level Tokenization)**

단어를 하위단어(Subword)나 글자(Character) 수준으로 분리하여 단어의 일부분만 토큰화하거나 글자 단위로 분리한다.
BERT와 같은 모델의 하위단어 토큰화에 나오는 "##"는 서브워드 토큰을 표시하기 위함이다.
Bert는 단어를 부분 단위로 쪼개어 희귀하거나 새로운 단어도 부분적으로 표현할 수 있도록 하고, 이를 통해 어휘 크기를 줄이고 다양한 언어 패턴을 학습할 수 있게 된다.
즉, 첫 단어를 기준으로 다양한 단어를 조합해 문맥을 파악하기 위함이다.

['un', '##ha', '##pp', '##iness']

위 결과의 un은 단어의 시작 부분으로 접두사 없이 표기된다.
'##ha', '##pp', '##iness'는 각각 단어의 중간과 끝 부분에 해당하므로 ##으로 시작한다.

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

text = 'unhappiness'
tokens = tokenizer.tokenize(text)
tokens
```

### WordPunctTokenizer와 TreebankWordTokenizer 비교

```python
from nltk.tokenize import WordPunctTokenizer, TreebankWordTokenizer

corpus = """
COVID-19(전염병), Dr.Smith(의사), NASA(우주항공국)와 같은 특정 기관이나 명칭이 있다. 특수 문자 또한 €100.50와 같은 가격을 나타내거나, 2023/10/30와 같은 날짜를 표현하는 데 사용될 수 있다. 이러한 경우, €100.50를 하나의 토큰으로 유지해야 할 필요가 있다.
"""

tokenizer = WordPunctTokenizer()                # nltk 저수준 토크나이저
tokens = tokenizer.tokenize(corpus)
print(tokens)

tokens = word_tokenize(corpus)                  # nltk 고수준 토크나이저 (내부적으로 Treebanktokenizer 와 유사한 방식 사용)
print(tokens)

tokenizer = TreebankWordTokenizer()             # nltk 저수준 토크나이저
tokens = tokenizer.tokenize(corpus)
print(tokens)
```

## 토큰화에 따른 감정분석 비교

* WordPunctTokenizer 토큰화
  구두점과 아포스트로피를 단어의 일부로 처리하지 않고 별도의 토큰으로 분리한다.
  예를 들어, Don't는 ["Don", "'", "t"]로 분리된다.
  이는 각 단어의 의미를 분리하는데 중점을 두고 있어 단어의 맥락이 약해질 수 있다.

* word_tokenize 토큰화 (TreebankWordTokenizer)
  아포스트로피와 하이픈을 포함한 단어를 하나의 토큰으로 처리한다.
  예를 들어, Don't는 ["Do", "n't"]로 아포스트로피를 분리하지만, 감정 분석에서 Don't는 부정적인 감정으로 해석될 수 있다.

> WordPunctTokenizer는 "Don't"를 ['Don', "'", 't']로 분리하여, 부정적인 감정이 제대로 전달되지 않게 된다. 이로 인해 "Don't"의 부정적인 의미가 상실되고, 감정 분석 결과가 왜곡될 수 있다. 또한 "well-being"을 ['well', '-', 'being']으로 분리하게 되어 긍정적인 의미를 제대로 반영하지 못할 수 있다.

### SentimentIntensityAnalyzer

영어만 처리 가능

SentimentIntensityAnalyzer는 NLTK의 VADER(Valence Aware Dictionary and sEntiment Reasoner) 감성 분석 도구의 핵심 클래스이다. VADER는 특히 소셜 미디어 텍스트처럼 짧고 구어체가 많은 문장에 최적화된 룰 기반(Lexicon-based) 감성 분석 방법이다.

| 키        | 설명              | 범위          |
| -------- | --------------- | ----------- |
| neg      | 부정적인 감정 비율      | 0.0 ~ 1.0   |
| neu      | 중립적인 감정 비율      | 0.0 ~ 1.0   |
| pos      | 긍정적인 감정 비율      | 0.0 ~ 1.0   |
| compound | 전체적인 감정을 종합한 지표 | -1.0 ~ +1.0 |

작동 과정 단계별 설명

1. 감성 사전(Lexicon) 기반 분석

   * VADER는 약 7,500여 개의 영어 단어와 각 단어의 감성 점수(긍정, 부정, 중립)를 포함한 감성 어휘 사전을 사용한다.
   * 각 단어는 [-4, +4] 범위의 감성 점수를 가진다. 예를 들어, "excellent"는 +3.1, "terrible"은 -2.5 등으로 미리 정의되어 있다.

2. 텍스트 전처리

   * 입력된 문장을 공백, 구두점, 대소문자 등에 따라 토큰화한다.
   * 감정 분석에 영향을 주는 이모티콘, 대문자, 느낌표, 부정어(not, never 등), 강조(very, extremely 등)도 함께 처리한다.

3. 단어별 감성 점수 계산

   * 각 토큰(단어 또는 기호)이 감성 사전에 있으면 해당 점수를 가져온다.
   * 부정어가 단어 앞에 오면 감성 점수를 반전시키거나 줄인다.
   * 강조(대문자, 느낌표, 강조 부사 등)가 있으면 감성 점수를 높인다.

4. 문장 전체 감성 점수 합산 및 정규화

   * 모든 단어의 감성 점수를 합산한다.
   * 이 합산 점수를 [-1, 1] 범위로 정규화하여 최종 감성 점수(compound score)를 산출한다.
   * 이 과정에서 문장의 길이, 강조, 부정어 등의 효과도 함께 반영된다.

5. 결과 반환

   * polarity_scores() 함수는 다음과 같은 네 가지 값을 반환한다:

     * 'neg': 부정 감성 점수 (0~1)
     * 'neu': 중립 감성 점수 (0~1)
     * 'pos': 긍정 감성 점수 (0~1)
     * 'compound': 전체 감성의 종합 점수 ([-1, 1] 범위, 0에 가까우면 중립, 1에 가까우면 긍정, -1에 가까우면 부정)

임계치(Threshold) 활용

* 일반적으로 compound 점수가 0.05 이상이면 긍정, -0.05 이하이면 부정, 그 사이면 중립으로 분류한다.

정리

* SentimentIntensityAnalyzer는 감성 사전과 규칙(룰)을 바탕으로 단어, 문장 구조, 강조, 부정어 등을 종합적으로 고려하여 텍스트의 감성 점수를 계산한다.
* 머신러닝 기반이 아니라 룰 기반이므로 빠르고 직관적이며, 특히 소셜 미디어 데이터에 적합하다.

추가 설명

* VADER는 영어에 최적화되어 있으며, 한글 등 다른 언어에는 별도의 감성 사전이 필요하다.
* 복잡한 문맥이나 아이러니, 반어법에는 한계가 있을 수 있다.

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

analyzer.polarity_scores("I love this product! It's amazing!")
```

```python
texts = [
    "I absolutely love this!",
    "This is okay, i guess",
    "I hate it so much"
]

for text in texts:
    print(text, "->", analyzer.polarity_scores(text))
```

```python
# 토큰화 결과가 감성 분석에 끼치는 영향
corpus = "I don't like the way she talks about mental well-being"

word_punct_tokenizer = WordPunctTokenizer()
tokens1 = word_punct_tokenizer.tokenize(corpus)
text1 = ' '.join(tokens1)
print(text1)

tokens2 = word_tokenize(corpus)
text2 = ' '.join(tokens2)
print(text2)

# 감성 분석
print(analyzer.polarity_scores(text1))
print(analyzer.polarity_scores(text2))
```

## 한글 문장 토큰화

### 한국어 토큰화의 특수성

한국어는 영어와 달리 교착어라서 단순 띄어쓰기만으로 정확한 단어를 나누는 것이 쉽지 않은데 이는 영어는 줄임말에 대한 예외처리를 한다면 띄어쓰기(whitespace)를 기준으로 하는 띄어쓰기 토큰화를 수행해도 단어 토큰화가 잘 작동한다.

하지만 한국어는 영어와 달리 교착어라서 단순 띄어쓰기만으로는 정확한 단어를 나누기 어렵다. 예를 들어 "그녀가 나를 사랑한다"에서 "그녀"와 "가"를 구분하는 것이 중요하다. 이를 위해서 형태소(morpheme) 분석이 필요하다.

### 교착어 특성

한국어는 영어와 달리 조사라는 특징이 있어, 같은 단어에도 다양한 조사가 붙어 사용된다.

예를 들어, "그"라는 단어는 "그가", "그에게", "그를", "그와", "그는" 등으로 변형될 수 있으며, 이러한 조사의 존재는 자연어 처리를 복잡하게 만들고, 동일한 단어임에도 서로 다른 조사가 붙으면 서로 다른 단어로 인식될 위험이 있다.

따라서 대부분의 한국어 NLP에서는 조사를 분리하는 과정이 필요하다.

한국어는 띄어쓰기 단위가 영어처럼 독립적인 단어로 구성되지 않기 때문에, 교착어의 특성을 이해하고 형태소(morpheme) 분석을 통해 정확한 단어 토큰화를 수행해야한다.

* 형태소 분석(Morpheme Analysis)

  * 형태소는 뜻을 가진 가장 작은 단위로, 자립 형태소와 의존 형태소로 나뉜다.
  * 자립 형태소는 명사나 동사처럼 독립적인 의미를 가지는 단어이고, 의존 형태소는 조사, 접미사 같은 단어에 붙어서 쓰이는 단어이다.

    * 자립 형태소: '책', '학교', '사람'
    * 의존 형태소: '-가', '-을', '-일', '-었', '-다'

* 한국어 띄어쓰기의 특성

  * 한국어는 띄어쓰기를 잘 지키지 않는 경향이 있다.
  * 이는 한국어가 띄어쓰기가 없어도 의미를 쉽게 이해할 수 있는 언어적 특성 때문이다. 반면 영어는 띄어쓰기가 없으면 문장의 의미를 파악하기 어려운 경우가 많다.
  * 예를 들어, "제가이렇게띄어쓰기를전혀하지않고글을썼다"고 하더라도 한국어는 이해할 수 있다.
  * 영어는 "tobeornottobethatisthequestion" 같은 문장은 해석하기 어렵다.
  * 결론적으로, 한국어는 많은 코퍼스에서 띄어쓰기가 지켜지지 않아 자연어 처리의 어려움을 초래하고 있다. 이러한 특성을 고려하여 한국어 자연어 처리에서는 형태소 분석 및 적절한 토큰화 기법이 필수적이다.

## 품사 태깅(part-of-Speech Tagging)

자연어 처리(NLP)에서 각 단어에 해당하는 품사(명사, 동사, 형용사 등)를 식별하고 태그를 부여하는 과정이다. 이 작업은 문장에서 단어의 의미와 문법적 역할을 이해하는 데 중요한 역할을 한다.

예시

* "Bat": 박쥐(명사), 배트(명사), 치다(동사)
* "배": 과일(명사), 신체(명사), 선박(명사)

위와 같이 한 단어가 여러 의미를 가지고 있으며 이는 문맥에 따라 의미가 달라지게 되며, 품사 태깅을 통해 어떤 품사로 사용되는지를 인식해야 올바른 의미를 파악할 수 있다.

**8품사**

1. 명사(Noun):
   사람, 장소, 사물 등을 지칭하는 말을 명사라고 한다. 형태와 기능에 따라 고유명사, 추상명사, 물질명사, 집합명사, 보통명사로 구분할 수 있다. 예: book, snow, train, paper, ship.

2. 대명사(Pronoun):
   명사를 대신하여 사용하는 말을 대명사라고 한다. 명사의 반복을 피하기 위해 주로 사용하며, 명사의 기능을 대신한다. 종류로는 인칭대명사, 지시대명사, 의문대명사, 부정대명사, 관계대명사가 있다. 예: I, you, he, she, they, we, it, that.

3. 동사(Verb):
   사람 또는 사물의 움직임(동작)이나 상태(입장, 형편, 처지)를 나타내는 말을 동사라고 한다.

4. 형용사(Adjective):
   명사나 대명사를 수식하는 말을 형용사라고 한다. 명사와 대명사의 모양, 크기, 색깔, 성질, 특색 등을 표현한다.

5. 부사(Adverb):
   형용사, 동사, 부사, 또는 문장 전체를 꾸미는 말을 부사라고 한다. 시간, 장소, 방법 등을 나타낼 때 사용한다.

6. 전치사(Preposition):
   명사나 대명사 앞에 놓여 다른 말과의 관계를 나타내는 말을 전치사라고 한다. 또한 자동사 뒤에 놓여 타동사로 만드는 역할을 하기도 하며, 글과 글을 연결하는 역할도 한다.

7. 접속사(Conjunction):
   두 개의 절이나 구, 또는 단어를 연결할 때 사용하는 말을 접속사라고 한다.

8. 감탄사(Interjection):
   감정을 나타낼 때 사용하는 말을 감탄사라고 한다. 예: oh!, wow!, oops!

### 구(phrase)와 절(clause)

두 개 이상의 단어가 모여 주어 + 동사의 관계가 있으면 절(clause)이고, 주어 + 동사의 관계가 없으면 구(phrase)이다. 이러한 구와 절은 하나의 품사처럼 역할을 한다.

1. 구(phrase): 주어 + 동사 관계가 없는 경우

구는 주어와 동사의 관계가 없는 단어들의 모임이다.

* 명사구(Noun Phrase):
  명사처럼 사용되는 구를 명사구라고 한다.
  예: My best friend is kind. (주어 역할), I met my best friend. (목적어 역할)

* 형용사구(Adjective Phrase):
  형용사처럼 사용되는 구를 형용사구라고 한다.
  예: He is not a man to be trusted.

* 부사구(Adverb Phrase):
  부사처럼 사용되는 구를 부사구라고 한다.
  예: Our school is on a hill.

2. 절(clause): 주어 + 동사 관계가 있는 경우

절은 주어와 동사의 관계가 있는 단어들의 모임이다.

* 명사절(Noun Clause):
  명사처럼 사용되는 절을 명사절이라고 한다.
  예: I know that you ate it.

* 형용사절(Adjective Clause):
  형용사처럼 사용되는 절을 형용사절이라고 한다.
  예: The girl that he loved very much left him.

* 부사절(Adverb Clause):
  부사처럼 사용되는 절을 부사절이라고 한다.
  예: He turned red when she smiled at him.

### pos_tag
part-of-speech 품사 태깅

pos_tag는 자연어 처리(NLP)에서 단어에 품사를 태깅하는 함수로, 주로 NLTK(Natural Language Toolkit)와 같은 라이브러리에서 사용된다.

**주요 품사 태깅**

| **태그**  | **설명**                                       | **예시**                               |
|-----------|------------------------------------------------|----------------------------------------|
| **NN**    | 단수 명사를 나타냄                              | "cat", "book", "apple"                |
| **NNS**   | 복수 명사를 나타냄                              | "cats", "books", "apples"             |
| **NNP**   | 단수 고유 명사를 나타냄                        | "Alice", "London", "NASA"             |
| **NNPS**  | 복수 고유 명사를 나타냄                        | "Smiths", "United Nations"            |
| **VB**    | 동사의 원형을 나타냄                           | "run", "eat", "play"                  |
| **VBP**   | 동사의 현재형 (1인칭, 2인칭, 복수 등) | "run", "eat", "play" |
| **VBZ**   | 3인칭 단수 현재형 동사를 나타냄                | "runs", "eats", "plays"               |
| **VBD**   | 동사의 과거형을 나타냄                         | "ran", "ate", "played"                |
| **VBG**   | 동명사 또는 현재 분사를 나타냄                 | "running", "eating", "playing"        |
| **VBN**   | 동사의 과거 분사형을 나타냄                    | "run" (as in "has run"), "eaten"      |
| **JJ**    | 형용사를 나타냄                                | "big", "blue", "happy"                |
| **JJR**   | 비교급 형용사를 나타냄                         | "bigger", "bluer", "happier"          |
| **JJS**   | 최상급 형용사를 나타냄                         | "biggest", "bluest", "happiest"       |
| **RB**    | 부사를 나타냄                                  | "quickly", "very", "well"             |
| **RBR**   | 비교급 부사를 나타냄                           | "more quickly", "better"              |
| **RBS**   | 최상급 부사를 나타냄                           | "most quickly", "best"                |
| **IN**    | 전치사 또는 종속 접속사를 나타냄               | "in", "on", "because"                 |
| **DT**    | 한정사를 나타냄                                | "the", "a", "some"                    |
| **PRP**   | 인칭 대명사를 나타냄                           | "I", "you", "he", "they"              |
| **PRP$**  | 소유 대명사를 나타냄                           | "my", "your", "his", "their"          |

```python
from nltk.tag import pos_tag

text = "I love hamburger!"

tokens = word_tokenize(text)
pos_tags = pos_tag(tokens)
pos_tags
```
