# 정규 표현식
**정규 표현식(Regular Expression)** 은 특정한 규칙을 가진 문자열을 찾기 위한 패턴이다.<br> 정규 표현식을 사용하면 대량의 텍스트 데이터에서 특정 패턴을 효율적으로 추출, 삭제, 대체할 수 있다. <br> 특히, 텍스트 전처리나 데이터 정제 작업에서 매우 유용하게 사용된다. <br>
이러한 정규 표현식은 특히 비밀번호 패턴 및 이메일 패턴과 같은 곳에서 자주 사용된다. <br>



> 정규 표현식 기본 문법

| 특수 문자 | 설명 | 예시 |
|---|---|---|
| . | 임의의 한 문자 | `a.c` : `abc`, `a1c` 등과 매치 |
| ? | 앞 문자가 0개 또는 1개 있을 때 매치 | `ab?c` : `abc`, `ac`와 매치 |
| * | 앞 문자가 0개 이상 있을 때 매치 | `ab*c` : `ac`, `abc`, `abbc` |
| + | 앞 문자가 1개 이상 있을 때 매치 | `ab+c` : `abc`, `abbc` |
| ^ | 문자열이 특정 문자로 시작할 때 매치 | `^abc` : `abcde`, `abc`와 매치 |
| $ | 문자열이 특정 문자로 끝날 때 매치 | `abc$` : `deabc`, `abc`와 매치 |
| {n} | 문자가 정확히 n번 반복될 때 매치 | `a{2}b` : `aab`와 매치 |
| {n,m} | 문자가 n번 이상 m번 이하 반복될 때 매치 | `a{2,4}b` : `aab`, `aaab`, `aaaab` |
| [ ] | 대괄호 안의 문자 중 하나와 매치 | `[abc]` : `a`, `b`, `c` |
| [^ ] | 대괄호 안의 문자 제외하고 매치 | `[^abc]` : `d`, `e`, `1` |
| \| | OR 연산자로 둘 중 하나와 매치 | `a\|b` : `a` 또는 `b` |


```python
import re
```

## 정규표현식 메소드

* `re.match()`: 시작부분 검증
* `re.search()`: 존재여부 확인(첫번째 결과 반환)
* `re.findall()`: 모든 결과 확인
* `re.finditer()`: 모든 결과 Match객체로 반환
* `re.sub()`: 치환
* `re.compile()`: 정규표현식객체 컴파일(r문자열) 재사용 효과적

## 정규표현식 문법

### .
임의의 한 글자

```python
pattern = re.compile('a.c')
print(pattern,type(pattern))

print(pattern.search('abc'))
print(pattern.search('aBc'))
print(pattern.search('a c'))
print(pattern.search('ac'))
print(pattern.search('I love abc!'))
```

### ? 수량자
`문자?` 문자가 0개 또는 1개

```python
pattern = re.compile('ab?c')

print(pattern.search('abc'))
print(pattern.search('ac'))
print(pattern.search('abbc'))
```

### \* 수량자
`문자*` 문자가 *0개이상*

```python
pattern = re.compile('ab*c')

print(pattern.search('ac'))
print(pattern.search('abc'))
print(pattern.search('abbc'))
print(pattern.search('aXc'))
```

```python
pattern = re.compile('a.*c')

print(pattern.search('aㅋㅋㅋc'))
print(pattern.search('ac'))
print(pattern.search('a123456c'))
print(pattern.search('aXc'))
```

### \+ 수량자
`문자+` 문자가 1개이상

```python
pattern = re.compile('ab+c')

print(pattern.search('abc'))
print(pattern.search('abbbc'))
print(pattern.search('ac'))
```

### {n} {min,max} 수량자
- `문자{n}` 문자가 n개
- `문자{min, max}` 문자가 min ~ max개
- `문자{min,}` 문자가 min ~ 개


```python
pattern = re.compile('ab{2}c')

print(pattern.search('abc'))
print(pattern.search('abbc'))
print(pattern.search('abbbc'))
print(pattern.search('abbbbc'))
```

```python
pattern = re.compile('ab{2,3}c')

print(pattern.search('abc'))
print(pattern.search('abbc'))
print(pattern.search('abbbc'))
print(pattern.search('abbbbc'))
```

```python
pattern = re.compile('ab{2,}c')

print(pattern.search('abc'))
print(pattern.search('abbc'))
print(pattern.search('abbbc'))
print(pattern.search('abbbbc'))
```

```python
pattern = re.compile('ab{,3}c')

print(pattern.search('abc'))
print(pattern.search('abbc'))
print(pattern.search('abbbc'))
print(pattern.search('abbbbc'))
```

### [ ]
문자하나에 대한 목록/범위 설정

```python
src = "ABCDEFGhijklmn abc 가나다라마바사ㅋㅋㅎㅎㅓㅓㅏㅏ 1234567890 !@#$%^&*() \t\n"

print(re.findall(r'[aj]',src))

print(re.findall(r'[0123456789]',src))
print(re.findall(r'[0-9]',src))

print(re.findall(r'[a-z]',src))
print(re.findall(r'[A-Z]',src))
print(re.findall(r'[A-Za-z]',src))
print(re.findall(r'[A-z]',src))
print(re.findall(r'[A-z]',src))

print(re.findall(r'[가-힣]',src))
print(re.findall(r'[가-힣ㄱ-ㅎㅏ-ㅣ]',src))

print(re.findall(r'[^가-힣]',src))
```

```python
pattern = re.compile('[0-9]{3}-[0-9]{3,4}-[0-9]{4}')

print(pattern.search('제 번호는 010-1234-5678입니다.'))
```

### 기본문자 클래스

| 기호   | 의미         | 포함 문자            | 예시 (매치됨)           | 예시 (매치 안 됨)    |
| ---- | ---------- | ---------------- | ------------------ | -------------- |
| `\w` | 단어 문자      | `[a-zA-Z0-9_]`   | `A`, `z`, `3`, `_` | `-`, `@`, `공백` |
| `\W` | **비단어 문자** | `[^a-zA-Z0-9_]`  | `-`, `@`, `공백`     | `A`, `3`, `_`  |
| `\d` | 숫자         | `[0-9]`          | `0`, `7`           | `a`, `@`, ` `  |
| `\D` | **숫자가 아님** | `[^0-9]`         | `A`, ` `, `!`      | `3`, `9`       |
| `\s` | 공백 문자      | `[ \t\n\r\f\v]`  | `공백`, `\t`, `\n`   | `A`, `3`, `-`  |
| `\S` | **공백이 아님** | `[^ \t\n\r\f\v]` | `A`, `3`, `-`      | `공백`, `\n`     |

```python
text = 'ID_007 is available.\nID_007은 사용가능합니다.'

print(re.findall(r'\w+',text))          # 유니코드 환경에서는 한글도 포함
print(re.findall(r'\d+',text))
print(re.findall(r'\s+',text))
```

```python
print(re.findall(r'\W+',text))          # 유니코드 환경에서는 한글도 포함
print(re.findall(r'\D+',text))
print(re.findall(r'\S+',text))
```

### ^ $ 앵커문자

- ^ 시작 매칭
- $ 종료 매칭

```python
pattern = re.compile('^abc')

print(pattern.search('abcㅋㅋㅋ'))
print(pattern.search('ㅋㅋㅋabc'))
print(re.search('^abc','ㅋㅋㅋ\nabc'))
print(re.search('^abc','ㅋㅋㅋ\nabc'),re.MULTILINE)
```

```python
print(re.search('world$','Hello world'))
print(re.search('world$','Hello world!'))
print(re.search('world$','world\ncome'))
print(re.search('world$','world\ncome',re.MULTILINE))
```

### |
or연산자

```python
pattern = re.compile("on|ues|rida")

text = "Monday Tuesday Wednesday Thursday Friday Saturday Sunday"

print(re.findall(f"on|ues|rida",text))
print(re.findall(f"(?:Mon|Tues|Fri)day",text))
```

### 그룹캡쳐

```python
text = "This is <strong>important</strong> text"

pattern = r"<strong>(.*)</strong>"

# sub(pattern, repl, string): string에서 pattern을 찾아 repl로 대체
print(re.sub(pattern,r'\1',text))       # \1, \2, ... 부여된 그룹 번호 호출
```

```python
text = "<div>Hello <b>World</b>! Visit <a href='https://example.com'>here</a>.</div>"

pattern = r"<(\w+)[^>]*>(.*)</\1>"

while re.search(pattern,text):
    text = re.sub(pattern,r'\2',text)
    print(text)

print(text)
```

## Flag
정규표현식의 작동방식을 제어하는 문법

**Python `re` 모듈 플래그 정리**

| 플래그 이름          | 축약어    | 설명                                                |
| --------------- | ------ | ------------------------------------------------- |
| `re.IGNORECASE` | `re.I` | 대소문자 구분 없이 매칭함 (`a`와 `A` 동일하게 취급)                 |
| `re.MULTILINE`  | `re.M` | `^`, `$`가 문자열 전체가 아닌 **각 줄의 시작/끝**에도 반응하게 만듦      |
| `re.DOTALL`     | `re.S` | `.`(아무글자)에 **줄바꿈 문자(`\n`)도 포함**해서 매칭되도록 함               |

```python
pattern = re.compile(r"hello",re.IGNORECASE)
print(pattern.search("HELLO WORLD"))

pattern_str = r'hello'
print(re.search(pattern_str,"HELLO WORLD",re.IGNORECASE))
```

## 정규표현식 토큰화

```python
from nltk.tokenize import RegexpTokenizer

text = "He's a runner, but not a long-distance runner."

tokenizer = RegexpTokenizer(r'[\w]+')
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
# 구분자 정규식 작성 gaps = True
tokenizer = RegexpTokenizer(r'\s+',gaps = True)
tokens = tokenizer.tokenize(text)
tokens
```

## 정규표현식을 이용한 정제

```python
# 이메일 제거
text = "문의: john.doe@example.com, 회사: contact@company.org"
text = re.sub(r'\S+@\S+','',text)
text
```

```python
# 공백과 단어문자를 제외한 특수 문자 제거
text = "Hello, welcome to NLP! Let's clean text :)"
text = re.sub(r"[^\s\w]",'',text)
text
```
